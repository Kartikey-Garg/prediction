<H1>Remove the trailing space from the symptom columns</H1>

```python
cols = df.columns
data = df[cols].values.flatten()

s = pd.Series(data)
s = s.str.strip()
s = s.values.reshape(df.shape)

df = pd.DataFrame(s, columns=df.columns)
df.head()
```

<H1>Fill the NaN values with zero</H1>

```python
df = df.fillna(0)
df.head()
```

<H1>Symptom severity rank</H1>

```python
df1 = pd.read_csv('Symptom-severity.csv')
df1.head()
```

<H1>Get overall list of symptoms<H1>

```python
df1['Symptom'].unique()
```

```python
vals = df.values
symptoms = df1['Symptom'].unique()

for i in range(len(symptoms)):
    vals[vals == symptoms[i]] = df1[df1['Symptom'] == symptoms[i]]['weight'].values[0]
    
d = pd.DataFrame(vals, columns=cols)
d.head()
```

```python
d = d.replace('dischromic _patches', 0)
d = d.replace('spotting_ urination',0)
df = d.replace('foul_smell_of urine',0)
df.head()
```

```python
(df[cols] == 0).all()
```

```python
df['Disease'].value_counts()
```

```python
df['Disease'].unique()
```

```python
data = df.iloc[:,1:].values
labels = df['Disease'].values
```

<H1>Split the data, labels with shuffle</H1>

```python
x_train, x_test, y_train, y_test = train_test_split(data, labels, shuffle=True, train_size = 0.85)
print(x_train.shape, x_test.shape, y_train.shape, y_test.shape)
```

<H1>Initialize and train a Support vector classifier</H1>

```python
model = SVC()
model.fit(x_train, y_train)
```

**Prediction**

```python
preds = model.predict(x_test)
```

```python
conf_mat = confusion_matrix(y_test, preds)
df_cm = pd.DataFrame(conf_mat, index=df['Disease'].unique(), columns=df['Disease'].unique())
print('F1-score% =', f1_score(y_test, preds, average='macro')*100, '|', 'Accuracy% =', accuracy_score(y_test, preds)*100)
sns.heatmap(df_cm)
```
